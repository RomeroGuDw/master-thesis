```python
import numpy as np
import matplotlib.pyplot as plt
import json
plt.rcParams["figure.facecolor"] = 'white'
root_path = 'C:/Users/David Romero/source/repos/Thesis/Thesis/'
```

## Scene Classification ##

```python
json_filename = 'secLyrSC_RF_it2_varImportance.json'
with open(root_path + json_filename ) as f:
    all_data = json.load(f)
    
data = np.array(all_data["opencv_ml_rtrees"]["var_importance"])

x = [i for i in range(1,len(data)+1)]

plt.figure()
plt.plot(x, data, linewidth=0.5,label='_nolegend_')

plt.axvspan(1, 512, facecolor='b', alpha=0.1)
plt.axvspan(513, 520, facecolor='g', alpha=0.1)
plt.axvspan(521, 1544, facecolor='r', alpha=0.1)

plt.legend((r'$\Psi_{i}(X)$','SC','OR'))

plt.xlim(0, len(x)+2)
plt.rcParams['axes.spines.top'] = False
plt.rcParams['axes.spines.right'] = False

plt.xlabel('Variable Index')

plt.title(r'Variable Importance - $2^{nd}$ layer feature vector - SC')

#plt.savefig('SC_importance_all.png', format='png', dpi=1000)

plt.show()
```

```python
class_names = ['','coast', 'forest', 'highway', 'insidecity', 'mountain', 'opencountry', 'street', 'tallbuilding','']

plt.figure()
n = 512
plt.matshow([data[n-1:n+8+1],data[n-1:n+8+1]])

tick_marks = np.arange(len(class_names))
plt.xticks(tick_marks, class_names, rotation=-0, horizontalalignment = 'center')
plt.yticks([])
plt.gca().xaxis.tick_bottom()
plt.title(r'Importance of the $1^{st}$ layer SC responses in the SC task')

plt.colorbar()

#plt.savefig('SC_importance_SC.png', format='png', dpi=1000)
plt.show()
```

```python
data_heatmap = []
for i in range(0,4):
    data_heatmap.append(data[(n + 8 + 256*i):(n + 8 + 256*i)+256])
    data_heatmap[i] = np.reshape(data_heatmap[i], (16,16))
    
for i in range(1,4):
    data_heatmap[0] = np.append(data_heatmap[0], data_heatmap[i], axis = 1)
    
plt.figure()
plt.matshow(data_heatmap[0])

[plt.plot([16*i + 15.5,16*i + 15.5],[-.5,15.5],color='w', linestyle='-', linewidth=2) for i in range(0,3)]

l1 = np.array((7, 16.5))
plt.text(l1[0], l1[1], 'car', fontsize=12)
plt.text(14.5+l1[0], l1[1], 'person', fontsize=12)
plt.text(16+15+l1[0], l1[1], 'horse', fontsize=12)
plt.text(16+16+15+l1[0], l1[1], 'cow', fontsize=12)

plt.colorbar()
plt.axis('off')
plt.title('Importance of the Detection Maps (OR) in the the SC task')

#plt.savefig('SC_importance_OR.png', format='png', dpi=1000)

plt.show()
```

## Object Recognition ##

```python
json_filename = 'secLyrOR_1RF_it2_varImportance.json'
with open(root_path + json_filename ) as f:
    all_data = json.load(f)
    
data = np.array(all_data["opencv_ml_rtrees"]["var_importance"])

x = [i for i in range(1,len(data)+1)]

plt.figure()
plt.plot(x, data, linewidth=0.5,label='_nolegend_')

plt.axvspan(1, 1765, facecolor='b', alpha=0.1)
plt.axvspan(1765, 1773, facecolor='g', alpha=0.1)
plt.axvspan(1773, len(data), facecolor='r', alpha=0.1)

plt.legend((r'$\Psi_{i}(X)$','SC','OR'))

plt.xlim(0, len(x)+2)
plt.rcParams['axes.spines.top'] = False
plt.rcParams['axes.spines.right'] = False

plt.title(r'Variable Importance - $2^{nd}$ layer feature vector - OR$_{car}$')
plt.xlabel('Variable Index')

plt.savefig('ORcar_importance_all.png', format='png', dpi=1000, bbox_inches="tight")

plt.show()
```

```python
class_names = ['coast', 'forest', 'highway', 'insidecity', 'mountain', 'opencountry', 'street', 'tallbuilding']

plt.figure()
n = 1765
plt.matshow([data[n:n+8],data[n:n+8]])

tick_marks = np.arange(len(class_names))
plt.xticks(tick_marks, class_names, rotation=-0, horizontalalignment = 'center')
plt.yticks([])

plt.gca().xaxis.tick_bottom()
plt.title(r'Importance of the $1^{st}$ layer SC responses in the  OR$_{car}$ task')

plt.colorbar()

plt.savefig('ORcar_importance_SC.png', format='png', dpi=1000, bbox_inches="tight")
plt.show()
```

```python
data_heatmap = []
for i in range(0,4):
    data_heatmap.append(data[(n + 8 + 256*i):(n + 8 + 256*i)+256])
    data_heatmap[i] = np.reshape(data_heatmap[i], (16,16))
    
for i in range(1,4):
    data_heatmap[0] = np.append(data_heatmap[0], data_heatmap[i], axis = 1)
    
plt.figure()
plt.matshow(data_heatmap[0])

[plt.plot([16*i + 15.5,16*i + 15.5],[-.5,15.5],color='w', linestyle='-', linewidth=2) for i in range(0,3)]

l1 = np.array((7, 16.5))
plt.text(l1[0], l1[1], 'car', fontsize=12)
plt.text(14.5+l1[0], l1[1], 'person', fontsize=12)
plt.text(16+15+l1[0], l1[1], 'horse', fontsize=12)
plt.text(16+16+15+l1[0], l1[1], 'cow', fontsize=12)

plt.colorbar()
plt.axis('off')
plt.title(r'Importance of the Detection Maps (OR) in the the OR$_{car}$ Task')
plt.savefig('ORcar_importance_OR.png', format='png', dpi=1000, bbox_inches="tight")
plt.show()
```
